4. Drop the columns that have more than 90% missing values. 
        - define a list columns_to_drop containing the column names to drop 
        - Create a new dataframe ebola_droppedCol
        - display the shape of the new dataframe.
        - print the number of missing values in ebola_droppedCol
        

```python
# Dropping the columns that have more than 90% missing values
columns_to_drop = []

for i in missing_val_count_by_column.index:
    percentage_missing = (missing_val_count_by_column[i] / len(ebola)) * 100
    if percentage_missing > 90:
        columns_to_drop.append(i)

ebola_droppedCol = ebola.drop(columns= columns_to_drop)

print(ebola_droppedCol.shape)

print(f'Number of missing values in ebola_droppedCol: {ebola_droppedCol.isnull().sum().sum()}')
print(ebola_droppedCol.isnull().sum())
```

#### Fill Backward

5. Answer each of the following questions: 
    - Count the number of missing values in the dataframe after the fill backward imputation of missing values. show the first 10 rows and 5 columns after imputation. 
    - Explain what changes have been made after fill backward imputation of the missing values and why. 


```python
ebola.iloc[0:10, 0:6]
```

```python
print(f"Initially null values:", ebola.isnull().sum().sum())

ebola_fill_b = ebola.fillna(method='bfill')

# Counting the number of missing values
result1 = ebola_fill_b.isnull().sum().sum()
print(f"Missing values after fill backward:",result1)
print(ebola_fill_b.isnull().sum())

ebola_fill_b.iloc[0:10, 0:6]
```

After fill backward imputation of the missing values, the null (NaN) values from the ebola dataset were replaced with the values of the rows below it if valid. There are still some missing values as there are consecutive null values from the bottom of the ebola dataset. This method is useful when the valid values are a reasonable estimate to replace the missing values.

```python
ebola.iloc[0:10, 0:6]
```

```python
print(f"Initially null values:", ebola.isnull().sum().sum())

ebola_fill_f = ebola.fillna(method='ffill')

# Counting the number of missing values
result1 = ebola_fill_f.isnull().sum().sum()
print(f"Missing values after fill forward:",result1)
print(ebola_fill_f.isnull().sum())

ebola_fill_f.iloc[0:10, 0:6]
```

Fill forward replaces the null values with the previous valid value from the head of the dataset. There are still some missing values as there are consecutive null values from the head of the ebola dataset. This method is useful when the valid values are a reasonable estimate to replace the missing values.

9. Write a paragraph to highlight the differences between the fill foward/backward versus linear interpolation of the missing values. 



Whilst fill forward/backward replace missing values in a dataset based on the next or previous valid value in the sequence, linear interpolation computes missing values by drawing a straight line between adjacent known data points to predict a reasonable value for the missing datapoints.

10. Time series interpolation works by calling interpolate with the option method='time'.
    - Explain how is it different than the linear interpolation? 
    - when do linear and time interpolation give the same results? 

Unlike linear interpolation which treats the values in the dataset as equally spaced, the time interpolation works with time series data that have a datetime type as index, and the actual time difference between the data points are taken into consideration when estimating the missing values.

The linear and time interpolation methods would give the same results if there is a datetime index that is equally spaced on a dataset.
